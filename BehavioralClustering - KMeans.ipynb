```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dataset = pd.read_excel("ExcelForClustering.xlsx")
X=Dataset.iloc[:,1:7]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=Dataset.iloc[:,1:7]
scaled_DataSet=scaler.fit_transform(X)
Y=scaled_DataSet
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(scaled_DataSet)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Based on the above elbow plot, we go for 4 Clusters

```python
#Building the KMeans Cluster
k=4
clusters=KMeans(k,random_state=42)
clusters.fit(scaled_DataSet)
Dataset["clusterid"]=clusters.labels_
Dataset
```
